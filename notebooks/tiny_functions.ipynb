```python
import pyclesperanto_prototype as cle
from skimage.io import imread, imshow

image = imread("C:/structure/data/blobs.tif")
imshow(image)
```

```python
from skimage.filters import threshold_otsu
from skimage.measure import label

def otsu_labeling(image):
    """
    Thresholds an image using the Otsu (1979, https://doi.org/10.1109/TSMC.1979.4310076) 
    technique and connected component labeling to segment objects
    """
    # binarize
    binary = image > threshold_otsu(image)
    
    # label
    return label(binary)
```

```python
segmented = otsu_labeling(image)

cle.imshow(segmented, labels=True, colorbar=True)
```
